```python
import numpy as np
import matplotlib.pyplot as plt

# konstanter
G = 6.67430e-11
c = 299792458
M = 1.989e30  # mass of the Sun
r_s = 2 * G * M / c**2  # Schwarzschild radius

# klot koordinater
theta = np.linspace(0, 2 * np.pi, 200)
phi = np.linspace(0, np.pi, 200)
theta, phi = np.meshgrid(theta, phi)

# Radial coordinate varies slightly across the surface (just for visualization)
r_min = 2.5 * r_s
r_max = 4.5 * r_s
r = r_min + (r_max - r_min) * np.sin(phi)  # vary radius from pole to equator

# kordinaterna
x = r * np.sin(phi) * np.cos(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(phi)

# berkna g00 som en funktion av r
g_00 = -(1 - r_s / r)

# normalisera fr frgkarta
norm_field = (g_00 - g_00.min()) / (g_00.max() - g_00.min())

# plottar
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, facecolors=plt.cm.inferno(norm_field), linewidth=0, antialiased=False)

ax.set_title("vartiation av g00 p en klot runt solen")
ax.set_xlabel("X (m)")
ax.set_ylabel("Y (m)")
ax.set_zlabel("Z (m)")
plt.colorbar(surface, shrink=0.5, aspect=10, label='g (normalized)')
plt.show()
```
